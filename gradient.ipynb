```python
import numpy as np
```

```python
path = '../dataset'
X_ = np.load(f'{path}/train/arr_X.npy').reshape(-1,1800)
Y_ = np.load(f'{path}/train/arr_Y.npy')
```

```python
def calculate_derivative(data):
    """
    1차원 데이터의 미분값을 계산하는 함수
    
    Parameters:
        data (numpy.ndarray): 1차원 시계열 데이터
    
    Returns:
        numpy.ndarray: 미분값으로 계산된 데이터 (원본 데이터 크기보다 1 작음)
    """
    # 이웃한 두 데이터 간의 차이를 계산하여 미분값 구하기
    derivative = np.diff(data)
    
    return derivative
```

```python
X_diff = calculate_derivative(X_)
```
